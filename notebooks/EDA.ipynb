# Exploratory Data Analysis: Student Performance

The goal of this notebook is to perform Exploratory Data Analysis (EDA) on the Student Exam Score predictor dataset. We'll examine the distributions of scores, look for missing values, and analyze how different demographic factors influence student performance.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

print("Libraries imported successfully!")
```

## 1. Load Data and Overview
Let's start by loading the data and taking a high-level look at its structure. For this demonstration, we'll mimic the real student performance dataset structure since we don't have the original `csv` connected.

```python
# Load actual data if available
# df = pd.read_csv('../data/raw/student_data.csv')

# For demonstration purposes, we'll create a mockup dataset mimicking the real expected schema
np.random.seed(42)
n_samples = 1000

data = {
    'gender': np.random.choice(['female', 'male'], n_samples),
    'race/ethnicity': np.random.choice(['group A', 'group B', 'group C', 'group D', 'group E'], n_samples),
    'parental level of education': np.random.choice(["bachelor's degree", 'some college', "master's degree", "associate's degree", 'high school', 'some high school'], n_samples),
    'lunch': np.random.choice(['standard', 'free/reduced'], n_samples),
    'test preparation course': np.random.choice(['none', 'completed'], n_samples),
    'math score': np.random.normal(66, 15, n_samples).clip(0, 100).astype(int),
    'reading score': np.random.normal(69, 14, n_samples).clip(0, 100).astype(int),
    'writing score': np.random.normal(68, 15, n_samples).clip(0, 100).astype(int)
}

df = pd.DataFrame(data)

# Inject a couple missing values to demonstrate our missing-values functionality block.
df.loc[10:25, 'math score'] = np.nan
df.loc[50:60, 'reading score'] = np.nan

print(f"Dataset Shape: {df.shape}")
display(df.head())
```

## 2. Missing Value Analysis
Missing values can skew algorithms. We analyze the volume of nulls using pandas and a visualization heatmap to map out where they commonly occur.

```python
missing_counts = df.isnull().sum()
missing_pct = (missing_counts / len(df)) * 100

missing_df = pd.DataFrame({'Missing Values': missing_counts, 'Percentage (%)': missing_pct})
display(missing_df[missing_df['Missing Values'] > 0])

plt.figure(figsize=(8, 4))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Value Matrix')
plt.show()
```

## 3. Distribution Plots of Scores
Visualizing target distributions is critical to verify if scores are normally distributed across students. If there's high skew, we might need a transform step in our preprocessing phase.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(df['math score'].dropna(), kde=True, color='skyblue', ax=axes[0])
axes[0].set_title('Math Score Distribution')

sns.histplot(df['reading score'].dropna(), kde=True, color='olive', ax=axes[1])
axes[1].set_title('Reading Score Distribution')

sns.histplot(df['writing score'].dropna(), kde=True, color='gold', ax=axes[2])
axes[2].set_title('Writing Score Distribution')

plt.tight_layout()
plt.show()
```

## 4. Correlation Heatmap
This checks how strictly Math, Reading, and Writing correlate to each other. For example, if reading score highly correlates to writing score, providing both variables to a Linear model could introduce multi-collinearity without dimensionality reduction.

```python
plt.figure(figsize=(6, 5))
corr_matrix = df[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=0, vmax=1, fmt='.2f')
plt.title('Correlation Heatmap of Exam Scores')
plt.show()
```

## 5. Comparison Plots
Here we compare different demographics against their respective scores using Box Plots. This is the simplest way to visualize shifts in the Median and quartiles between classes.

```python
# Gender vs Scores
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
score_cols = ['math score', 'reading score', 'writing score']

for i, col in enumerate(score_cols):
    sns.boxplot(data=df, x='gender', y=col, ax=axes[i], palette='Set2')
    axes[i].set_title(f'Gender vs {col.title()}')

plt.tight_layout()
plt.show()
```

```python
# Parental Education vs Scores
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

for i, col in enumerate(score_cols):
    sns.boxplot(data=df, x='parental level of education', y=col, ax=axes[i], palette='Pastel1', order=["master's degree", "bachelor's degree", "associate's degree", 'some college', 'high school', 'some high school'])
    axes[i].set_title(f'Parental Education vs {col.title()}')
    axes[i].tick_params(axis='x', rotation=15)

plt.tight_layout()
plt.show()
```
